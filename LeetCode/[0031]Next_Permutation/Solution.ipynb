```python
class Solution:
    def nextPermutation(self, nums):
        """
        :type nums: List[int]
        :rtype: void Do not return anything, modify nums in-place instead.
        """
        pivot=-1
        
        i=len(nums)-1
        while i > 0:
            if nums[i-1]<nums[i]:
                pivot=i-1
                break
            i-=1

        if pivot==-1:
            nums[:]=nums[::-1]
            return
        
        i=pivot+1
        while i<len(nums):
            if nums[i]>nums[pivot] and (i==len(nums)-1 or nums[i+1]<=nums[pivot]):
                nums[i],nums[pivot]=nums[pivot],nums[i]
                nums[pivot+1:len(nums)]=reversed(nums[pivot+1:len(nums)])
                return
            i+=1
```

```python
s=Solution()
nums=[5,4,7,5,3,2]
s.nextPermutation(nums)
print(nums)
```
